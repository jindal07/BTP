```python
import requests
from parser import parse_table

BASE_URL = "https://www.nirfindia.org/Rankings/{year}/EngineeringRanking.html"

def scrape_nirf_data(year):
    """Scrapes NIRF data for a given year and returns a DataFrame."""
    url = BASE_URL.format(year=year)
    try:
        response = requests.get(url, timeout=15)
        if response.status_code != 200:
            print(f" Failed to fetch {year} page: {response.status_code}")
            return None

        html_content = response.text
        df = parse_table(html_content, year)
        return df

    except Exception as e:
        print(f" Scraping error for {year}: {e}")
        return None
```
