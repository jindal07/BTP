```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv("csv_data/nirf_combined_data.csv")

for year in range(2017, 2025):
    col = f'Rank_{year}'
    if col in df.columns:
        df[col] = df[col].astype(str).str.extract(r'(\d+)').astype(float)

long_df = pd.DataFrame()
for year in range(2017, 2026):
    temp_df = pd.DataFrame({
        'Institute ID': df['Institute ID'],
        'Name': df['Name'],
        'Year': year,
        'Rank': df.get(f'Rank_{year}'),
        'Score': df.get(f'Score_{year}'),
        'TLR': df.get(f'TLR (100)_{year}'),
        'RPC': df.get(f'RPC (100)_{year}'),
        'GO': df.get(f'GO (100)_{year}'),
        'OI': df.get(f'OI (100)_{year}'),
        'Perception': df.get(f'PERCEPTION (100)_{year}')
    })
    long_df = pd.concat([long_df, temp_df], ignore_index=True)

long_df.dropna(inplace=True)

train_years = list(range(2017, 2025))
year_weights = {year: 1.0 - (2024 - year) * 0.1 for year in train_years}
long_df['Weight'] = long_df['Year'].map(year_weights)

train_df = long_df[long_df['Year'].isin(train_years)]
test_df = long_df[long_df['Year'] == 2025]

features = ['Score', 'TLR', 'RPC', 'GO', 'OI', 'Perception']

X_train = train_df[features]
y_train = (
    train_df['TLR'] * 0.3 +
    train_df['RPC'] * 0.3 +
    train_df['GO'] * 0.2 +
    train_df['OI'] * 0.1 +
    train_df['Perception'] * 0.1
)
w_train = train_df['Weight']

X_test = test_df[features]
y_test = (
    test_df['TLR'] * 0.3 +
    test_df['RPC'] * 0.3 +
    test_df['GO'] * 0.2 +
    test_df['OI'] * 0.1 +
    test_df['Perception'] * 0.1
)
college_names_test = test_df['Name']

model = XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=5, random_state=42)
model.fit(X_train, y_train, sample_weight=w_train)

y_pred = model.predict(X_test)

results_df = pd.DataFrame({
    'College': college_names_test.values,
    'Actual_Score_2025': y_test.values,
    'Predicted_Score_2025': y_pred
})
results_df['Predicted_Rank_2025'] = results_df['Predicted_Score_2025'].rank(ascending=False, method='min').astype(int)
results_df['Actual_Rank_2025'] = results_df['Actual_Score_2025'].rank(ascending=False, method='min').astype(int)
results_df.sort_values(by='Predicted_Rank_2025', inplace=True)
results_df.reset_index(drop=True, inplace=True)
results_df.index = range(1, len(results_df) + 1)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("RÂ² Score:", r2)
print(results_df.head(10))

results_df.to_csv("nirf_predictions_2025.csv", index=False)

sns.set(style="whitegrid", font_scale=1.1)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=results_df['Actual_Score_2025'], y=results_df['Predicted_Score_2025'])
plt.plot([60, 95], [60, 95], 'r--')
plt.xlabel("Actual Score (2025)")
plt.ylabel("Predicted Score (2025)")
plt.title("Predicted vs Actual NIRF Scores (2025)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x=results_df['Actual_Rank_2025'], y=results_df['Predicted_Rank_2025'])
plt.plot([0, 100], [0, 100], 'r--')
plt.xlabel("Actual Rank (2025)")
plt.ylabel("Predicted Rank (2025)")
plt.title("Predicted vs Actual NIRF Ranks (2025)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,6))
score_error = results_df['Predicted_Score_2025'] - results_df['Actual_Score_2025']
sns.histplot(score_error, bins=30, kde=True, color='purple')
plt.title("Distribution of Score Prediction Error (2025)")
plt.xlabel("Score Error")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,6))
rank_error = results_df['Predicted_Rank_2025'] - results_df['Actual_Rank_2025']
sns.histplot(rank_error, bins=20, kde=True, color='orange')
plt.title("Distribution of Rank Prediction Error (2025)")
plt.xlabel("Rank Error")
plt.tight_layout()
plt.show()
```
