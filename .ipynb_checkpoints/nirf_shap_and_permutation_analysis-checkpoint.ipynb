```python
import pandas as pd
import shap
from sklearn.inspection import permutation_importance
from xgboost import XGBRegressor
import matplotlib.pyplot as plt

df = pd.read_csv("csv_data/nirf_combined_data.csv")

long_df = pd.DataFrame()
for year in range(2017, 2024):
    temp_df = pd.DataFrame({
        'Institute ID': df['Institute ID'],
        'Name': df['Name'],
        'Year': year,
        'Score': df.get(f'Score_{year}'),
        'TLR': df.get(f'TLR (100)_{year}'),
        'RPC': df.get(f'RPC (100)_{year}'),
        'GO': df.get(f'GO (100)_{year}'),
        'OI': df.get(f'OI (100)_{year}'),
        'Perception': df.get(f'PERCEPTION (100)_{year}')
    })
    long_df = pd.concat([long_df, temp_df], ignore_index=True)

long_df.dropna(inplace=True)

features = ['TLR', 'RPC', 'GO', 'OI', 'Perception']
target = 'Score'  

train_df = long_df[long_df['Year'] < 2024].dropna(subset=features + [target])
test_df = long_df[long_df['Year'] == 2024].dropna(subset=features)

X_train = train_df[features]
y_train = train_df[target]

model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)
model.fit(X_train, y_train)

explainer = shap.Explainer(model, X_train)
shap_values = explainer(X_train)

shap.summary_plot(shap_values, X_train, plot_type="bar")
plt.show()

perm_importance = permutation_importance(model, X_train, y_train, n_repeats=10, random_state=42)
sorted_idx = perm_importance.importances_mean.argsort()

plt.barh([features[i] for i in sorted_idx], perm_importance.importances_mean[sorted_idx])
plt.xlabel("Permutation Importance")
plt.title("Feature Importance (Permutation)")
plt.show()
```
